```python
with open('mydata/test_data.txt', 'r') as f:
    content = f.readlines()
content
```

```python
tokenizer = get_tokenizer('basic_english')

def yield_tokens(examples):
    for example in examples:
        tokens = tokenizer(example)
        yield tokens
token_generator = yield_tokens(content)
# list(token_generator)
```

```python
vocab = build_vocab_from_iterator(token_generator)
# vocab.get_stoi()
```
